```python
import yfinance as yf
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

def check_stationarity(series, name="Series", signif=0.05, verbose=True):
    """
    Performs the Augmented Dickey-Fuller test for stationarity.
    """
    if verbose:
        print(f"--- ADF Test for {name} ---")
    try:
        # Ensure series is not all NaN or too short
        series_dropna = series.dropna()
        if series_dropna.empty:
            if verbose: print(f"ADF Test failed for {name}: Series is all NaN after dropna.")
            print("-" * 30)
            return False
        # A common heuristic for minimum sample size for ADF is related to T^(1/3)
        # For simplicity, we'll use a slightly more generous check than just a fixed number for very short series
        min_obs = max(20, int(4 * (len(series_dropna)**(2/9))) if len(series_dropna) > 0 else 20) # Ensure min_obs is reasonable
        if len(series_dropna) < min_obs:
            if verbose: print(f"ADF Test failed for {name}: Series too short ({len(series_dropna)} points) after dropna. Min ~{min_obs} recommended.")
            print("-" * 30)
            return False

        result = adfuller(series_dropna, autolag='AIC')
        if verbose:
            print(f'ADF Statistic: {result[0]}')
            print(f'p-value: {result[1]}')
            print('Critical Values:')
            for key, value in result[4].items():
                print(f'\t{key}: {value}')

        is_stationary = result[1] <= signif
        if verbose:
            if is_stationary:
                print(f"Conclusion: Series is likely STATIONARY (p-value <= {signif}). Reject H0.")
            else:
                print(f"Conclusion: Series is likely NON-STATIONARY (p-value > {signif}). Fail to reject H0.")
            print("-" * 30)
        return is_stationary
    except Exception as e:
        if verbose:
            print(f"ADF Test failed for {name}: {e}")
            print("-" * 30)
        return False # Assume non-stationary or problematic if test fails

def analyze_asset_relationship(symbol1, symbol2, start_date, end_date, interval='1d',
                               asset_type='generic', significance_level=0.05, verbose_plots=True):
    """
    Calculates and prints correlation and cointegration for two assets.
    Asset type can be 'forex', 'futures', or 'generic' (for stocks).
    """
    print(f"\n\n==========================================================")
    print(f"Analyzing {symbol1} and {symbol2} | Interval: {interval}")
    print(f"Period: {start_date} to {end_date}")
    print(f"==========================================================")

    # 1. Construct Tickers and Download Data
    ticker1_req, ticker2_req = symbol1, symbol2
    if asset_type == 'forex':
        ticker1_req = f"{symbol1}=X"
        ticker2_req = f"{symbol2}=X"
    elif asset_type == 'futures':
        ticker1_req = f"{symbol1}=F"
        ticker2_req = f"{symbol2}=F"

    symbols_req_list = [ticker1_req, ticker2_req]
    try:
        data = yf.download(symbols_req_list, start=start_date, end=end_date, interval=interval, progress=False, auto_adjust=True, back_adjust=True)

        if data.empty:
            print(f"No data downloaded for {symbols_req_list} with interval {interval}. Check symbols, date range, and interval limits.")
            return None, None, False

        prices = data.get('Close', pd.DataFrame()) # Use .get for safer access
        if prices.empty:
             print(f"No 'Close' price data found in downloaded data for {symbols_req_list}.")
             return None, None, False

        prices = prices.dropna()

        if prices.empty or len(prices) < 50:
            print(f"Not enough valid data points ({len(prices)}) after cleaning for {symbols_req_list}. Min 50 required. Aborting.")
            return None, None, False

        # Robust column name matching (yfinance can sometimes return slightly different names)
        s1_col, s2_col = None, None
        for col_option in [ticker1_req, symbol1, ticker1_req.split('=')[0], symbol1.split('=')[0]]:
            if col_option in prices.columns:
                s1_col = col_option
                break
        for col_option in [ticker2_req, symbol2, ticker2_req.split('=')[0], symbol2.split('=')[0]]:
            if col_option in prices.columns:
                s2_col = col_option
                break

        if not s1_col and len(prices.columns) == 2: # If only two columns, assume they are the ones
            s1_col = prices.columns[0]
        if not s2_col and len(prices.columns) == 2:
            s2_col = prices.columns[1]
        if s1_col == s2_col and len(prices.columns) > 1 : # If they resolved to same and more than one col exists, try to pick distinct
            if prices.columns[0] != s1_col : s1_col = prices.columns[0]
            if prices.columns[1] != s2_col : s2_col = prices.columns[1]


        if not s1_col or not s2_col or s1_col == s2_col:
            print(f"Could not reliably identify price columns for {symbol1} ({ticker1_req}) or {symbol2} ({ticker2_req}).")
            print(f"Attempted s1_col: {s1_col}, s2_col: {s2_col}. Available columns: {list(prices.columns)}")
            return None, None, False

        series1 = prices[s1_col]
        series2 = prices[s2_col]
        # Rename series to original symbols for consistency in plots/outputs if names were changed
        series1.name = symbol1
        series2.name = symbol2


    except Exception as e:
        print(f"Error downloading or processing data for {symbols_req_list}: {e}")
        return None, None, False

    # 2. Check individual series for stationarity
    print("\n--- Stationarity Check (ADF Test) ---")
    is_s1_stationary = check_stationarity(series1, symbol1, signif=significance_level)
    is_s2_stationary = check_stationarity(series2, symbol2, signif=significance_level)

    if is_s1_stationary or is_s2_stationary:
        print("\nWarning: At least one of the individual series appears to be stationary (I(0)).")
        print("Cointegration is typically sought between non-stationary (I(1)) series.")
        print("Proceeding, but interpret cointegration results with caution.\n")


    # 3. Calculate Correlation
    print("\n--- Correlation ---")
    try:
        # Reconstruct a temporary DataFrame with original symbol names for correlation matrix
        corr_prices_df = pd.DataFrame({symbol1: series1, symbol2: series2})
        correlation_matrix = corr_prices_df.corr()
        correlation = correlation_matrix.loc[symbol1, symbol2]
        print(f"Correlation between {symbol1} and {symbol2}: {correlation:.4f}")
        if verbose_plots:
            plt.figure(figsize=(6, 4))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
            plt.title(f'Correlation: {symbol1} & {symbol2} ({interval})')
            plt.show()
    except Exception as e:
        print(f"Could not calculate correlation: {e}")
        correlation = None


    # 4. Calculate Cointegration (Engle-Granger Test)
    print("\n--- Cointegration (Engle-Granger Test) ---")
    try:
        if len(series1.dropna()) < 20 or len(series2.dropna()) < 20:
            print("Series too short for cointegration test after NA drop.")
            return correlation, None, False

        coint_test_result = coint(series1, series2, trend='c', autolag='AIC')
        coint_t_statistic = coint_test_result[0]
        coint_p_value = coint_test_result[1]
        critical_values = coint_test_result[2]

        print(f"Cointegration t-statistic: {coint_t_statistic:.4f}")
        print(f"P-value: {coint_p_value:.4f}")
        print(f"Critical values (1%, 5%, 10%): {critical_values[0]:.4f}, {critical_values[1]:.4f}, {critical_values[2]:.4f}")

        is_cointegrated = coint_p_value < significance_level

        if is_cointegrated:
            print(f"\nConclusion: {symbol1} and {symbol2} ARE LIKELY COINTEGRATED (p-value < {significance_level}).")
            series2_with_const = sm.add_constant(series2) # X variable
            model = sm.OLS(series1, series2_with_const).fit() # Y ~ X

            beta_param_name = series2.name if series2.name in model.params.index else series2_with_const.columns[1]
            beta = model.params[beta_param_name]

            print(f"Estimated hedge ratio (beta for {symbol2} to predict {symbol1}): {beta:.4f}")
            print(f"Spread = {symbol1} - {beta:.4f} * {symbol2}")
            spread = series1 - beta * series2

            if verbose_plots:
                plt.figure(figsize=(12, 4))
                spread.plot(title=f'Cointegrating Spread: {symbol1} - {beta:.2f}*{symbol2} ({interval})')
                plt.axhline(spread.mean(), color='red', linestyle='--', label=f'Mean: {spread.mean():.4f}')
                plt.ylabel("Spread Value")
                plt.legend()
                plt.show()

                # Z-Score Plot (added from previous request)
                if not spread.empty:
                    spread_mean = spread.mean()
                    spread_std = spread.std()
                    if pd.notnull(spread_std) and spread_std != 0:
                        zscore = (spread - spread_mean) / spread_std
                        plt.figure(figsize=(12, 6))
                        zscore.plot(title=f'Z-Score of Spread ({symbol1} - {beta:.2f}*{symbol2})')
                        plt.axhline(0, color='black', linestyle='--', linewidth=0.8, label='Mean Spread (Z=0)')
                        plt.axhline(2.0, color='red', linestyle='--', linewidth=1.0, label='Z = +2.0 (Upper Threshold)')
                        plt.axhline(-2.0, color='green', linestyle='--', linewidth=1.0, label='Z = -2.0 (Lower Threshold)')
                        plt.axhline(1.0, color='salmon', linestyle=':', linewidth=0.8, label='Z = +1.0')
                        plt.axhline(-1.0, color='lightgreen', linestyle=':', linewidth=0.8, label='Z = -1.0')
                        plt.legend(loc='best')
                        plt.ylabel('Z-score')
                        plt.title(f'Z-Score of Spread: {symbol1} vs {symbol2} ({interval})')
                        plt.grid(True)
                        plt.show()
                    elif verbose_plots:
                         print("Spread standard deviation is zero or NaN. Cannot plot Z-score.")
                elif verbose_plots:
                     print("Spread is empty. Cannot plot Z-score.")

            print("\nChecking stationarity of the spread (ADF Test):")
            check_stationarity(spread, "Cointegrating Spread", signif=significance_level)
        else:
            print(f"\nConclusion: {symbol1} and {symbol2} ARE LIKELY NOT COINTEGRATED (p-value >= {significance_level}).")

    except Exception as e:
        print(f"Cointegration test or spread analysis failed: {e}")
        coint_p_value = None
        is_cointegrated = False

    # 5. Plot normalized prices
    if verbose_plots and not series1.empty and not series2.empty:
        print("\n--- Plotting Normalized Prices ---")
        plt.figure(figsize=(12, 6))
        try:
            (series1 / series1.iloc[0] * 100).plot(label=f"{symbol1} (Normalized)")
            (series2 / series2.iloc[0] * 100).plot(label=f"{symbol2} (Normalized)")
            plt.title(f"Normalized Prices: {symbol1} vs {symbol2} ({interval})")
            plt.ylabel("Normalized Price (Start=100)")
            plt.xlabel("Date/Time") # Changed from "Date"
            plt.legend()
            plt.grid(True)
            plt.show()
        except IndexError:
            print("Could not plot normalized prices due to insufficient data at start.")
        except Exception as e:
            print(f"Could not plot normalized prices: {e}")

    return correlation, coint_p_value, is_cointegrated


# --- Configuration ---
alpha = 0.05 # Significance level for tests

# Define candidate pairs: list of tuples (symbol1, symbol2, asset_type)
candidate_pairs = [
    ("ES", "NQ", 'futures'),  # E-mini S&P 500 vs E-mini NASDAQ 100
    ("ES", "YM", 'futures'),  # E-mini S&P 500 vs Mini-DOW
    ("ES", "RTY", 'futures'), # E-mini S&P 500 vs Russell 2000
    ("ES", "EMD", 'futures'), # E-mini S&P 500 vs E-mini Midcap 400
    ("NQ", "YM", 'futures'),  # E-mini NASDAQ 100 vs Mini-DOW
    ("NQ", "RTY", 'futures'), # E-mini NASDAQ 100 vs Russell 2000
    ("YM", "RTY", 'futures'),  # Mini-DOW vs Russell 2000

    # III. Precious Metals (Using standard contract sizes for each distinct metal)
    ("GC", "SI", 'futures'),  # Gold vs Silver
    ("GC", "PL", 'futures'),  # Gold vs Platinum
    ("SI", "PL", 'futures'),  # Silver vs Platinum
    # ("PL", "PA", 'futures'), # Platinum vs Palladium (often correlated) - PA is in your list

    # IV. Energy Complex (Using standard contract sizes for each distinct product)
    ("CL", "HO", 'futures'),  # Crude Oil vs Heating Oil
    ("CL", "RB", 'futures'),  # Crude Oil vs New York Harbor (RBOB) Gasoline
    ("HO", "RB", 'futures'),  # Heating Oil vs New York Harbor (RBOB) Gasoline
    # ("CL", "NG", 'futures'), # Crude Oil vs Natural Gas (can be correlated, but also distinct drivers)

    # V. Grains & Oilseeds (Soy Complex)
    ("ZS", "ZM", 'futures'),  # Soybeans vs Soybean Meal
    ("ZS", "ZL", 'futures'),  # Soybeans vs Soybean Oil
    ("ZM", "ZL", 'futures'),  # Soybean Meal vs Soybean Oil
    ("ZC", "ZW", 'futures'),  # Corn vs Wheat

    # VI. Livestock
    ("LE", "GF", 'futures'),  # Live Cattle vs Feeder Cattle

    # VII. Foreign Exchange (Currency Futures - distinct currency pairs)
    ("6A", "6N", 'futures'),  # Australian Dollar vs New Zealand Dollar
    ("6A", "6C", 'futures'),  # Australian Dollar vs Canadian Dollar
    ("6C", "6N", 'futures'),  # Canadian Dollar vs New Zealand Dollar
    ("6E", "6B", 'futures'),  # Euro FX vs British Pound
    ("6E", "6S", 'futures'),  # Euro FX vs Swiss Franc

]

# --- Run Analysis ---
results_summary = []

# --- CHOOSE YOUR TIMEFRAME AND PERIODS ---
# For 1-MINUTE analysis:
# yfinance provides 1m data for the last 7 calendar days.
# To get a full day, set start to the target day and end to the day after.
# `end_date` is exclusive for yfinance intraday data.
today_dt = datetime.now()

# Determine the most recent *full* trading day for which we want data.
# This logic tries to get the previous UTC day's data.
# Be mindful of market holidays and exact market open/close times for specific assets.
# If today is Monday, it will aim for Friday's data.
if today_dt.weekday() == 0: # Monday
    target_day_for_data = today_dt - timedelta(days=3) # Previous Friday
elif today_dt.weekday() == 6: # Sunday (markets likely closed, aim for Friday)
    target_day_for_data = today_dt - timedelta(days=2) # Previous Friday
else: # Other weekdays (including Saturday, aim for Friday)
    target_day_for_data = today_dt - timedelta(days=1) # Previous day (if Sat, this is Fri)

# Check if the target_day_for_data is too far back (older than 6 days from today, yfinance might not serve 1m)
# We want data FOR target_day_for_data, so yf.download start=target_day_for_data, end=target_day_for_data+1day
if (today_dt.date() - target_day_for_data.date()).days > 6 :
    print(f"Warning: Calculated target day ({target_day_for_data.strftime('%Y-%m-%d')}) is more than 6 days ago.")
    print("YFinance might not serve 1m data. Adjusting to a more recent day.")
    # Fallback: use 2 days ago as target start, so data for 1 day before that.
    target_day_for_data = today_dt - timedelta(days=8)


current_start_dt_obj = target_day_for_data
current_end_dt_obj = target_day_for_data + timedelta(days=1)

current_start = current_start_dt_obj.strftime('2025-05-05')
current_end = current_end_dt_obj.strftime('2025-05-13')

current_interval = '1m'
print(f"USING 1-MINUTE DATA: Attempting to fetch for the full day of {current_start} (data from {current_start} up to, but not including, {current_end})")


if __name__ == "__main__":
    for s1, s2, asset_cat in candidate_pairs:
        # For 1m data, verbose_plots=True can generate many plots.
        # Consider setting to False if running many pairs and only checking summary.
        corr, p_val, is_coint = analyze_asset_relationship(
            s1, s2,
            start_date=current_start,
            end_date=current_end,
            interval=current_interval,
            asset_type=asset_cat,
            significance_level=alpha,
            verbose_plots=True # Set to False to quickly screen many pairs
        )
        results_summary.append({
            "pair": f"{s1}-{s2}",
            "asset_type": asset_cat,
            "interval": current_interval,
            "start_date_analyzed": current_start, # Record actual date used
            "correlation": corr,
            "coint_p_value": p_val,
            "is_cointegrated": is_coint
        })

    print("\n\n--- Results Summary ---")
    summary_df = pd.DataFrame(results_summary)
    if not summary_df.empty:
        summary_df = summary_df.sort_values(by="coint_p_value", ascending=True)
        print(summary_df)

        cointegrated_pairs_df = summary_df[summary_df['is_cointegrated'] == True]
        if not cointegrated_pairs_df.empty:
            print("\n--- Cointegrated Pairs Found ---")
            print(cointegrated_pairs_df)
        else:
            print("\n--- No Cointegrated Pairs Found with current settings ---")
    else:
        print("\n--- No analysis results to summarize (likely data fetching issues for all pairs) ---")
```
