```python
import yfinance as yf
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

def calculate_and_plot_spread_characteristics(
    symbol1, symbol2, start_date, end_date,
    interval='1d', asset_type='generic', verbose_plots=True):
    """
    Calculates and plots the spread, its mean, std dev, and correlation for two assets.
    Uses a static beta calculated over the entire period for spread calculation.
    """
    print(f"\n\n{'='*60}")
    print(f"Analyzing Spread for: {symbol1} vs {symbol2}")
    print(f"Period: {start_date} to {end_date} | Interval: {interval}")
    print(f"{'='*60}")

    # 1. Construct Tickers and Download Data
    ticker1_req, ticker2_req = symbol1, symbol2
    if asset_type == 'forex':
        ticker1_req = f"{symbol1}=X"
        ticker2_req = f"{symbol2}=X"
    elif asset_type == 'futures':
        ticker1_req = f"{symbol1}=F"
        ticker2_req = f"{symbol2}=F"
    # For 'crypto' or 'generic', no suffix needed for yfinance typically

    symbols_req_list = [ticker1_req, ticker2_req]
    try:
        print(f"Downloading data for: {symbols_req_list}...")
        # Using auto_adjust=False and back_adjust=False and then selecting 'Adj Close' if available
        # provides more control, but for simplicity with futures/forex, let yf handle it.
        # auto_adjust=True is generally fine for 'Close' prices for non-stocks or if you don't need distinct Open/High/Low/Close.
        data = yf.download(symbols_req_list, start=start_date, end=end_date, interval=interval,
                           progress=False, auto_adjust=True, back_adjust=True) # Let yf handle adjustments
        print(f"Raw data downloaded: {len(data)} bars.")

        if data.empty:
            print("No data downloaded. Check symbols, date range, and interval.")
            return

        prices_df = data.get('Close')
        if prices_df is None or prices_df.empty:
            print("No 'Close' price data found.")
            return
        print(f"'Close' price data extracted: {len(prices_df)} bars.")

        # Drop rows where EITHER asset has a NaN for 'Close'
        initial_bars = len(prices_df)
        prices_df = prices_df.dropna()
        print(f"Price data after dropping rows with any NaN: {len(prices_df)} bars (dropped {initial_bars - len(prices_df)}).")

        if len(prices_df) < 20: # Need enough data for regression and stats
            print(f"Insufficient data ({len(prices_df)} bars) after cleaning. Min 20 required. Aborting.")
            return

        # --- Robust Column Name Matching ---
        s1_col, s2_col = None, None
        # Try exact ticker match first
        if ticker1_req in prices_df.columns: s1_col = ticker1_req
        if ticker2_req in prices_df.columns: s2_col = ticker2_req

        # Fallback for symbol name if ticker_req not found (e.g. crypto)
        if not s1_col and symbol1 in prices_df.columns: s1_col = symbol1
        if not s2_col and symbol2 in prices_df.columns: s2_col = symbol2

        # If still not found and only two columns exist, assume they are the ones
        if not s1_col and len(prices_df.columns) == 2: s1_col = prices_df.columns[0]
        if not s2_col and len(prices_df.columns) == 2:
            if prices_df.columns[1] != s1_col: # Ensure it's different
                s2_col = prices_df.columns[1]
            elif prices_df.columns[0] != s1_col: # Fallback if first guess for s1_col was wrong
                 s2_col = prices_df.columns[0]


        if not s1_col or not s2_col or s1_col == s2_col:
            print(f"Could not reliably identify distinct price columns for {symbol1} ({ticker1_req}) or {symbol2} ({ticker2_req}).")
            print(f"Identified s1_col: {s1_col}, s2_col: {s2_col}. Available columns: {list(prices_df.columns)}")
            return

        series1 = prices_df[s1_col].rename(symbol1) # Ensure series name is simple
        series2 = prices_df[s2_col].rename(symbol2) # Ensure series name is simple
        print(f"Using column '{s1_col}' for {symbol1} and '{s2_col}' for {symbol2}.")
        print(f"Data points for analysis: {len(series1)}")

    except Exception as e:
        print(f"Error during data download or processing: {e}")
        return

    # 2. Calculate Correlation
    print("\n--- Correlation ---")
    try:
        correlation = series1.corr(series2)
        print(f"Correlation between {symbol1} and {symbol2}: {correlation:.4f}")
        if verbose_plots:
            plt.figure(figsize=(8, 6))
            # Scatter plot with regression line
            sns.regplot(x=series2, y=series1, ci=None, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
            plt.title(f'Price Relationship & Regression: {symbol1} vs {symbol2}')
            plt.xlabel(f'{symbol2} Price')
            plt.ylabel(f'{symbol1} Price')
            plt.grid(True)
            plt.show()
    except Exception as e:
        print(f"Could not calculate or plot correlation: {e}")
        correlation = None # Keep going if possible

    # 3. Calculate Hedge Ratio (Beta) using OLS on the entire period
    # This beta is descriptive for this specific historical period.
    # For backtesting, beta must be calculated on a rolling/expanding formation period.
    print("\n--- Hedge Ratio (Beta) Calculation ---")
    try:
        series2_with_const = sm.add_constant(series2) # Independent variable (X)
        model = sm.OLS(series1, series2_with_const).fit() # Dependent variable (Y)
        beta = model.params[symbol2] # Get beta for series2 (symbol2 is now the column name)
        intercept = model.params['const']
        print(f"OLS Regression: {symbol1} = {beta:.4f} * {symbol2} + {intercept:.4f}")
        print(f"Calculated Beta (Hedge Ratio for {symbol2}): {beta:.4f}")
    except Exception as e:
        print(f"Could not calculate hedge ratio (beta): {e}")
        return # Cannot proceed without beta

    # 4. Calculate Spread
    spread = series1 - beta * series2
    spread.name = 'Spread'
    print("\n--- Spread Characteristics ---")
    spread_mean = spread.mean()
    spread_std = spread.std()
    print(f"Spread Mean: {spread_mean:.4f}")
    print(f"Spread Standard Deviation: {spread_std:.4f}")

    if pd.isna(spread_std) or spread_std == 0:
        print("Spread standard deviation is NaN or zero. Cannot plot Z-score or std dev bands meaningfully.")
        # Still plot spread if possible, but skip std dev bands and Z-score

    # 5. Plotting
    if verbose_plots:
        # Plot 1: Normalized Prices
        plt.figure(figsize=(12, 6))
        try:
            if not series1.empty and not series2.empty() and not pd.isna(series1.iloc[0]) and not pd.isna(series2.iloc[0]):
                (series1 / series1.iloc[0] * 100).plot(label=f"{symbol1} (Normalized)")
                (series2 / series2.iloc[0] * 100).plot(label=f"{symbol2} (Normalized)")
                plt.title(f"Normalized Prices: {symbol1} vs {symbol2} ({interval})")
                plt.ylabel("Normalized Price (Start=100)")
                plt.legend()
                plt.grid(True)
                plt.show()
            else:
                print("Skipping normalized price plot due to empty series or NaN at start.")
        except Exception as e:
            print(f"Error plotting normalized prices: {e}")


        # Plot 2: Spread with Mean and Std Dev Bands
        plt.figure(figsize=(12, 6))
        spread.plot(label='Spread', color='blue')
        plt.axhline(spread_mean, color='red', linestyle='--', label=f'Spread Mean ({spread_mean:.2f})')
        if pd.notna(spread_std) and spread_std != 0:
            plt.axhline(spread_mean + spread_std, color='green', linestyle=':', label=f'+1 Std Dev ({spread_mean + spread_std:.2f})')
            plt.axhline(spread_mean - spread_std, color='green', linestyle=':', label=f'-1 Std Dev ({spread_mean - spread_std:.2f})')
            plt.axhline(spread_mean + 2 * spread_std, color='orange', linestyle=':', label=f'+2 Std Dev ({spread_mean + 2 * spread_std:.2f})')
            plt.axhline(spread_mean - 2 * spread_std, color='orange', linestyle=':', label=f'-2 Std Dev ({spread_mean - 2 * spread_std:.2f})')
        plt.title(f'Spread ({symbol1} - {beta:.2f}*{symbol2}) with Mean & Std Dev Bands')
        plt.ylabel("Spread Value")
        plt.legend()
        plt.grid(True)
        plt.show()

        # Plot 3: Z-score of the Spread (if std is valid)
        if pd.notna(spread_std) and spread_std != 0:
            z_score_spread = (spread - spread_mean) / spread_std
            plt.figure(figsize=(12, 6))
            z_score_spread.plot(label='Spread Z-Score', color='purple')
            plt.axhline(0, color='black', linestyle='--')
            plt.axhline(1.0, color='lightgreen', linestyle=':')
            plt.axhline(-1.0, color='lightgreen', linestyle=':')
            plt.axhline(2.0, color='salmon', linestyle=':')
            plt.axhline(-2.0, color='salmon', linestyle=':')
            plt.title(f'Z-Score of Spread ({symbol1} - {beta:.2f}*{symbol2})')
            plt.ylabel("Z-Score")
            plt.legend()
            plt.grid(True)
            plt.show()
        else:
            print("Skipping Z-score plot as spread standard deviation is invalid.")

    print(f"{'-'*60}\n")


# --- Configuration ---
candidate_pairs = [
    ("ES", "NQ", 'futures'),
    ("ES", "YM", 'futures'),
    ("ES", "RTY", 'futures'),
    ("NQ", "YM", 'futures'),
    ("NQ", "RTY", 'futures'),
    ("YM", "RTY", 'futures'),
    ("EURUSD", "GBPUSD", 'forex'),
    ("BTC-USD", "ETH-USD", 'crypto') # For crypto, yfinance usually doesn't need "=X"
]

# --- Date Configuration (using the logic for recent data) ---
today_dt = datetime.now()
delta_days_for_target = 1
if today_dt.weekday() == 0: delta_days_for_target = 3 # Monday -> prev Friday
elif today_dt.weekday() == 6: delta_days_for_target = 2 # Sunday -> prev Friday
elif today_dt.weekday() == 5: delta_days_for_target = 1 # Saturday -> prev Friday

# Cap how far back we look for 1m data (yfinance limit ~7 days)
# If we want data FOR analysis_day_str, start=analysis_day_str, end=day_after_analysis_day_str
# So, analysis_day_str should be within ~6 days from today.
if delta_days_for_target > 6:
    print(f"Warning: Calculated target day ({delta_days_for_target} days ago) might be too old for 1m yfinance data.")
    delta_days_for_target = min(6, delta_days_for_target) # Cap it, but still allow going back a bit if needed

analysis_day_dt_obj = today_dt - timedelta(days=delta_days_for_target)
end_date_for_yf_dt_obj = analysis_day_dt_obj + timedelta(days=1) # yf 'end' is exclusive

analysis_start_date = analysis_day_dt_obj.strftime('2025-05-01')
analysis_end_date = end_date_for_yf_dt_obj.strftime('2025-05-19')

# --- CHOOSE ANALYSIS PARAMETERS ---
# For daily data over a longer period:
# current_interval_to_analyze = '1d'
# current_start_date_to_analyze = (today_dt - timedelta(days=365*1)).strftime('%Y-%m-%d') # Last 1 year
# current_end_date_to_analyze = today_dt.strftime('%Y-%m-%d')

# For 1-minute data for the most recent full trading day:
current_interval_to_analyze = '15m'
current_start_date_to_analyze = analysis_start_date
current_end_date_to_analyze = analysis_end_date # yf.download end is exclusive

print(f"Analyzing with interval: {current_interval_to_analyze}")
print(f"Data period for analysis: {current_start_date_to_analyze} to {current_end_date_to_analyze} (end date exclusive for yf.download)")


if __name__ == "__main__":
    for s1, s2, asset_cat in candidate_pairs:
        # For 'crypto', yfinance doesn't use an asset_type suffix like '=X' or '=F'
        # Pass 'generic' or handle it inside the function if 'crypto' has special yf ticker needs
        effective_asset_type = asset_cat
        if asset_cat == 'crypto':
            effective_asset_type = 'generic' # Or handle crypto tickers specifically if needed

        calculate_and_plot_spread_characteristics(
            s1, s2,
            start_date=current_start_date_to_analyze,
            end_date=current_end_date_to_analyze,
            interval=current_interval_to_analyze,
            asset_type=effective_asset_type,
            verbose_plots=True
        )
```
